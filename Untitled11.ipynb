```python
def transform_into_x_and_y(df):
    X = df.drop(['Adj Close'],axis= 1)
    y = df['Adj Close']
    return X,y
```

```python
mrf_X,mrf_y = transform_into_x_and_y(mrf_df)
coalIndia_X,coalIndia_y = transform_into_x_and_y(coalIndia_df)
mindtree_X,mindtree_y = transform_into_x_and_y(mindtree_df)
```

```python
def split_data(X,y,percent):
    xtrain = X[:int(X.shape[0]*percent)]
    xtest = X[int(X.shape[0]*percent):]
    ytrain = y[:int(X.shape[0]*percent)]

    ytest = y[int(X.shape[0]*percent):]
    return xtrain,xtest,ytrain,ytest
```

```python
mrf_xtrain,mrf_xtest,mrf_ytrain,mrf_ytest = split_data(mrf_X,mrf_y,0.8)
```

```python
mindtree_xtrain,mindtree_xtest,mindtree_ytrain,mindtree_ytest = split_data(mindtree_X,mindtree_y,0.8)
```

```python
coalIndia_xtrain,coalIndia_xtest,coalIndia_ytrain,coalIndia_ytest = split_data(coalIndia_X,coalIndia_y,0.8)
```

# Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
def pipeline_for_linear_model(xtrain,ytrain,model):
    pipe = Pipeline(steps=[('sscaler',StandardScaler()),('model',model)])
    df_pipe = pipe.fit(xtrain,ytrain)
    return df_pipe
```

```python
mrf_pipe_linear = pipeline_for_linear_model(mrf_xtrain,mrf_ytrain,LinearRegression())
coalIndia_pipe_linear = pipeline_for_linear_model(coalIndia_xtrain,coalIndia_ytrain,LinearRegression())
mindtree_pipe_linear = pipeline_for_linear_model(mindtree_xtrain,mindtree_ytrain,LinearRegression())
```

# MAE,MSE,RMSE

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
def errors(df_pipe,xtest,ytest):
    ypredict = df_pipe.predict(xtest)
    print('MAE score is ',mean_absolute_error(ytest,ypredict))
    print('MSE score is ',mean_squared_error(ytest,ypredict))
    print('RMSE score is ',np.sqrt(mean_squared_error(ytest,ypredict)))
    print('R2 score is ',df_pipe.score(xtest,ytest))
```

```python
print('Error in model of mrf')
errors(mrf_pipe_linear,mrf_xtest,mrf_ytest)
```

```python
print('Error in model of coal India')
errors(coalIndia_pipe_linear,coalIndia_xtest,coalIndia_ytest)
```

```python
print('Error in model of Mindtree')
errors(mindtree_pipe_linear,mindtree_xtest,mindtree_ytest)
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
param_grid1 = {'max_depth':[50,100,150,200,250],'max_leaf_nodes':[10,20,30,40,50]}
```

```python
def model_with_grid(pipe,xtrain,ytrain,param_grid1):
    grid = GridSearchCV(pipe,param_grid=param_grid1,cv=5)
    grid.fit(xtrain,ytrain)
    return grid
```

```python
mrf_grid_dtr = model_with_grid(DecisionTreeRegressor(random_state=0),mrf_xtrain,mrf_ytrain,param_grid1)
errors(mrf_grid_dtr,mrf_xtest,mrf_ytest)
```

```python
mindtree_grid_dtr = model_with_grid(DecisionTreeRegressor(random_state=0),mindtree_xtrain,mindtree_ytrain,param_grid1)
errors(mindtree_grid_dtr,mindtree_xtest,mindtree_ytest)
```

```python
coalIndia_grid_dtr = model_with_grid(DecisionTreeRegressor(random_state=0),coalIndia_xtrain,coalIndia_ytrain,param_grid1)
errors(coalIndia_grid_dtr,coalIndia_xtest,coalIndia_ytest)
```
